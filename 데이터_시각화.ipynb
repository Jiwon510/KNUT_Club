```python
# DataFrame 스캐터 차트 실행 - 다수의 스캐터 차트
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("/sample_data/sample.csv")

# 차트 윈도우 제목
plt.figure(num="성적 그래프")

# 차트 글꼴
plt.rcParams.update({'font.family':'NanumBarunGothic', 'font.size':12})

# 차트 제목
plt.title("학번별 성적")

# x,y축 제목
plt.xlabel("학번")
plt.ylabel("성적")

# 한 개의 화면에서 나타나도록 하기 위한 axes 설정
ax = plt.gca()

xcolumns = ['학번', '학번', '학번']
ycolumns = ['운영체제', '논리회로', 'DB']

# 색상
color = ['#209FDF', '#99CA53', '#F6A625', '#6D5FD5', '#BF593E', '#FF0000', '#0000FF']

# 스캐터 차트
data.plot(kind='scatter', x='학번', y='운영체제', ax=ax, color=color[0], s=10)
data.plot(kind='scatter', x='학번', y='논리회로', ax=ax, color=color[1], s=20)
data.plot(kind='scatter', x='학번', y='DB', ax=ax, color=color[2], s=30)

plt.show()
```

```python
# DataFrame 파이 차트 실행 
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({'mass':[0.330, 4.87, 5.97], 'radius':[2439.7, 6051.8, 6378.1]})
index = ['Mercury', 'Venus', 'Earth']

# 차트 윈도우 제목
plt.figure(num="파이 차트")

# 차트 글꼴
plt.rcParams.update({'font.family':'NanumBarunGothic', 'font.size':12})

# 차트 제목
plt.title("행성 크기 비교")

# 한 개의 화면에서 나타나도록 하기 위한 axes 설정
ax = plt.gca()

data.plot(kind='pie', y='mass', ax=ax)

plt.show()
```

```python
# DataFrame 히스토그램 실행
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("/sample_data/sample.csv")

# 학번열 삭제
data = data.drop('학번', axis=1)

# 차트 윈도우 제목
plt.figure(num="성적 그래프")

# 차트 글꼴
plt.rcParams.update({'font.family':'NanumBarunGothic', 'font.size':12})

# 차트 제목
plt.title("과목별 점수 빈도")

# 한 개의 화면에서 나타나도록 하기 위한 axes 설정
ax = plt.gca()

# 히스토그램
data.plot(kind='hist', by=None, bins=20, ax=ax, alpha=0.5, edgecolor='black', linewidth=1)

# x,y축 제목
plt.xlabel("과목별 성적")
plt.ylabel("빈도수")

plt.show()
```
